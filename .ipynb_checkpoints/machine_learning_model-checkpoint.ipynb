```python
import pandas as pd
import nltk
from textblob import TextBlob
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in all of the cleaned csvs and assign them to dataframes
genre_df = pd.read_csv("Resources/Genre_info.csv")
movie_df = pd.read_csv("Resources/Movie_info.csv")
review_df = pd.read_csv("Resources/Review_info.csv")
```

```python
#display the review_df
review_df
```

```python
#display movie_df
movie_df
```

```python
genre_df

cleaner_genre_df = genre_df.drop_duplicates()
cleaner_genre_df
```

```python
cleaner_movie_df = movie_df.drop_duplicates()
cleaner_review_df = review_df.drop_duplicates()

combined_movie = cleaner_review_df.merge(cleaner_movie_df, how= "inner", on= "movie_id")
combined_movie
```

```python
pos_neg_ind_2 = np.where(combined_movie["rating_ind"]<8,0, 1)
combined_movie["pos_neg_ind_2"] = pos_neg_ind_2.tolist()
```

```python
combined_movie
```

```python
pos_neg_count=combined_movie.groupby('pos_neg').count()
plt.bar(pos_neg_count.index.values, pos_neg_count['rating_text'])
plt.xlabel('Review Sentiments')
plt.ylabel('Number of Review')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import RegexpTokenizer

#remove symbols and numbers
token = RegexpTokenizer(r'[a-zA-Z0-9]+')
cv = CountVectorizer(lowercase=True,stop_words='english',ngram_range = (1,1),tokenizer = token.tokenize)
text_counts= cv.fit_transform(combined_movie['rating_text'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    text_counts, combined_movie['rating_text'], test_size=0.2, random_state=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf=TfidfVectorizer()
text_tf= tf.fit_transform(combined_movie['rating_text'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    text_tf, combined_movie['pos_neg_ind_2'], test_size=0.2, random_state=12)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

clf = MultinomialNB().fit(X_train, y_train)
predicted= clf.predict(X_test)
print("MultinomialNB Accuracy:",metrics.accuracy_score(y_test, predicted))
```
