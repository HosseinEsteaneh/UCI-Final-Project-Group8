```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.imdb.com/list/ls025152261/?sort=release_date,desc&st_dt=&mode=detail&page=1"
browser.visit(url)
```

```python
for x in range(1):
    # HTML object
    html = browser.html
    
    # Parse HTML
    soup = BeautifulSoup(html, 'html.parser')
    
    #Retrieve all elements that contain movie title info
    movie_list = soup.find_all("div",{'class':'lister-item-content'})
    
    print(movie_list)
```

movie_titles = soup.find_all('h3', class_='lister-item-header')

print(movie_titles)

```python
movie_titles = soup.find_all('h3', class_='lister-item-header')

print(movie_titles)
```

```python
movie_list = list(movie_titles)

df = pd.DataFrame(np.array(movie_list))

df.head(100)
```

```python
import numpy as np
import nltk
from nltk import sent_tokenize, word_tokenize, pos_tag
```

```python
#tokenizer

tokens = word_tokenize(str(movie_titles))
tokens
```

```python
len(tokens)
```

```python
tagged_tokens = pos_tag(tokens)
tagged_tokens
```

```python
for element in tokens:
    if(type(element) == str):
        print(element)
```

```python
#new empty list to store movie ids in or values with tt######
movie_ids = []

for element in tokens:
    if len(element) == 17:
        clean_element = (element[7:18])
        if clean_element == 'item-index':
            del(clean_element)
        else:
            movie_ids.append(clean_element[0:8])
```

```python
print(movie_ids)
```
